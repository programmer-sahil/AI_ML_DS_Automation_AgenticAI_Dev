```python
# String methods

a = "Madhumita"
print(a.upper())
print(a.lower())

b = "    Madhu is studying at IUBH.   "
print(b.strip()) # remove unnecessay spaces at the beginning & ending


a = "Hello, World!"
print(a.replace("World", "Madhu"))


c = "My name is Madhu!"
print(c.split(" ")) # convert string to list

```

```python
# String Concatenation

print("Madhu" + "Mita")
print("Madhu" + 20)

# you can't add string + int (you will get an error: TypeError)
```

```python
a = "Hello"
b = "World"
c = a + " " + b # string is denoted by "" or ''
print(c)
```

```python
price = 59
txt = f"The price is {price:.2f} dollars"
print(txt)
```

```python
# Slicing Strings

# index starts from zero (from left side)
# index starts from -1 (from right side)

b = "Hello, World!"
print(b[2:5]) # output: llo

c = "Hello, World!"
print(c[-1]) # !

d = "Hello, World!"
print(d[:5]) # Hello

e = "Hello, World!"
print(e[-5:-2]) # orl
```

```python
# Multiline Strings

a = """Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua."""
print(a)
```

```python
# Looping Through a String

for x in "banana":
  print(x)
```

```python
# len() function counts total elements of the string

a = "Madhu"
print(len(a))
```

```python
txt = "The best things in life are free!"
print("free" in txt) # membership operator: in, not int (used in for loop & other places)
print("Madhu" in txt)
```
